```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import colors
import pandas as pd
import numpy as np
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import ListedColormap
```

```python

# Function to read an Excel file and return a DataFrame
def read_excel_file(file_path):
    return pd.read_excel(file_path)

# List of file paths
file_paths = [
    '/content/drive/MyDrive/Classroom/2019-2024.xlsx',
    '/content/drive/MyDrive/Classroom/2015-2019.xlsx',
    '/content/drive/MyDrive/Classroom/2010-2014.xlsx',
    '/content/drive/MyDrive/Classroom/2005-2009.xlsx',
    '/content/drive/MyDrive/Classroom/2000-2004.xlsx'
]

# Using list comprehension to apply the function to all file paths
dfs = [read_excel_file(path) for path in file_paths]

# dfs is a list of DataFrames. Access them using dfs[0], dfs[1], etc.
```

```python
# Create a deep copy of the DataFrame
df_copy = combined_df.copy()

# List of columns to be dropped
columns_to_drop = [
    "DEPARTMENT OF TRANSPORTATION TIRE IDENTIFIER",
    "TIRE SIZE",
    "LOCATION OF TIRE CODE",
    "TYPE OF TIRE FAILURE CODE",
    "WAS PART ORIGINAL EQUIPMENT",
    "DATE OF MANUFACTURE",
    "TYPE OF CHILD SEAT CODE",
    "INSTALLATION SYSTEM CODE",
    "DEALER'S NAME",
    "DEALER'S TELEPHONE NUMBER",
    "DEALER'S CITY",
    "DEALER'S STATE CODE",
    "DEALER'S ZIPCODE",
    "PRODUCT TYPE CODE",
    "WAS DEFECTIVE TIRE REPAIRED",
    "WAS ORIGINAL OWNER",
    "VEHICLE'S VIN",
    "DATE PURCHASED",
    "NUMBER OF CYLINDERS",
    "DRIVE TRAIN TYPE",
    "FUEL SYSTEM CODE",
    "FUEL TYPE",
    "VEHICLE TRANSMISSION TYPE",
    "ODINO",
    "SOURCE OF COMPLAINT CODE",
    "DATE ADDED TO FILE",
    "DATE COMPLAINT RECEIVED BY NHTSA",
    "PRODUCT TYPE CODE",
    "MEDICAL ATTENTION REQUIRED",
    "NUMBER OF OCCURRENCES"
]


```

```python
# Drop the specified columns from the copy of the DataFrame
df_copy.drop(columns=columns_to_drop, axis=1, inplace=True)

```

```python
null_counts = df_copy.isnull().sum()

# To display the total number of null values for each column
print(null_counts)
```

```python
df_copy_1 = df_copy.copy()

# List of columns to check for null values
columns_to_check = [
    "MANUFACTURER'S NAME",
    "VEHICLE MAKE",
    "DATE OF INCIDENT",
    "VEHICLE MODEL",
    "MODEL YEAR",
    "CONSUMER'S STATE CODE",
    "SPECIFIC COMPONENT'S DESCRIPTION",
    "CONSUMER'S CITY",
    "DESCRIPTION OF THE COMPLAINT",
    "WAS INCIDENT REPORTED TO POLICE",
    "ANTI-LOCK BRAKES",
    "WAS VEHICLE TOWED",
    "CRUISE CONTROL"
]

# Drop rows where any of the specified columns have null values
df_copy_1.dropna(subset=columns_to_check, inplace=True)

```

```python
df_copy_1.isnull().sum()
```

```python
# Convert the 'DATE OF INCIDENT' to string, extract year and month, and assign to new columns
df_copy_1['Year'] = df_copy_1['DATE OF INCIDENT'].apply(lambda x: str(int(x))[:4] )

# drop the 'DATE OF INCIDENT' column as it's no longer needed
df_copy_1.drop('DATE OF INCIDENT', axis=1, inplace=True)

```

```python
# Display the first few rows to verify the changes
df_copy_1.head()
```

```python
df_copy_1['Year'] = pd.to_numeric(df_copy_1['Year'])
```

```python
# Boxplot to visualize outliers
plt.figure(figsize=(10, 2))
plt.boxplot(df_copy_1['Year'], vert=False)  # Drop NA values if any
plt.title('Boxplot of YEAR')
plt.xlabel('Year')
plt.show()
```

```python
rows_outside_range = df_copy_1.query('Year < 2004 or Year > 2024')
```

```python
unique_years_outside_range = rows_outside_range['Year'].unique()


unique_years_outside_range
```

```python

# Use df.query() to filter rows where 'Year' is between 2004 and 2024, inclusive
df = df_copy_1.query('2004 <= Year <= 2024')

# Display the shape or a sample of the filtered DataFrame to verify the results
print(df.shape)


df.head()
```
